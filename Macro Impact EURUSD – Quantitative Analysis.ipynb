```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from fredapi import Fred
from datetime import datetime

# Set plot style
plt.style.use('seaborn-v0_8')
sns.set_palette("deep")

# 1. Pull Data
# Initialize FRED API with provided API key
fred = Fred(api_key='e1020df616b278b14d10a28764a8b2a1')

# Define date range
start_date = '2006-01-01'
end_date = '2025-05-04'

# Pull EUR/USD from yfinance with error handling
try:
    eurusd = yf.download('EURUSD=X', start=start_date, end=end_date, progress=False)
    if eurusd.empty:
        raise ValueError("No EUR/USD data retrieved from yfinance.")
    # Handle MultiIndex and ensure EURUSD column
    close_data = eurusd['Close']
    print("Type of eurusd['Close']:", type(close_data))
    if isinstance(close_data, pd.Series):
        eurusd = close_data.to_frame().rename(columns={'Close': 'EURUSD'})
    elif isinstance(close_data, pd.DataFrame):
        eurusd = close_data.rename(columns={close_data.columns[0]: 'EURUSD'})
    else:
        raise ValueError(f"Unexpected type for eurusd['Close']: {type(close_data)}")
    print("EUR/USD Data Retrieved:")
    print(eurusd.head())
    print(eurusd.tail())
    print("EUR/USD Columns:", eurusd.columns.tolist())
except Exception as e:
    print(f"Error retrieving EUR/USD data: {e}")
    raise

# Pull macro indicators from FRED with flexible handling
def fetch_fred_series(series_id, name):
    data = fred.get_series(series_id, start_date=start_date, end_date=end_date)
    print(f"FRED {series_id} Type:", type(data))
    print(f"FRED {series_id} Head:", data.head())
    if isinstance(data, pd.Series):
        return data.to_frame(name=name)
    elif isinstance(data, pd.DataFrame):
        return data.rename(columns={data.columns[0]: name})
    else:
        raise ValueError(f"Unexpected data type for {series_id}: {type(data)}")

# Fetch FRED series
us10y = fetch_fred_series('DGS10', 'US10Y')
ffr = fetch_fred_series('FEDFUNDS', 'FFR')
cpi = fetch_fred_series('CPIAUCSL', 'CPI')
dxy = fetch_fred_series('DTWEXBGS', 'DXY')

# 2. Data Wrangling
# Combine all data into a single DataFrame
data = eurusd.join([us10y, ffr, cpi, dxy], how='outer')

# Forward fill missing values and drop any remaining NaNs
data = data.ffill().dropna()

# Check if EURUSD column exists
if 'EURUSD' not in data.columns:
    raise KeyError("EURUSD column missing in combined DataFrame. Check data retrieval or join operation.")

# Calculate daily returns for EUR/USD
data['EURUSD_Returns'] = data['EURUSD'].pct_change()

# Ensure data aligns with your validated shape (5111 rows, 5 columns + returns)
print("Data Shape:", data.shape)
print("Combined Data Tail:")
print(data.tail())

# 3. Analytics and Visualizations
# Time-series trend plot
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['EURUSD'], label='EUR/USD', color='blue')
plt.title('EUR/USD Trend (2006-2025)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('EUR/USD', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Rolling correlation (90-day window) between EUR/USD returns and macro indicators
# Create a DataFrame to store correlations
corr_data = pd.DataFrame(index=data.index)
for col in ['US10Y', 'FFR', 'CPI', 'DXY']:
    corr_data[col] = data['EURUSD_Returns'].rolling(window=90).corr(data[col])
corr_data = corr_data.dropna()
corr_data['Date'] = corr_data.index
print("Correlation Data Shape:", corr_data.shape)
print("Correlation Data Columns:", corr_data.columns.tolist())
print("Correlation Data Head:")
print(corr_data.head())
print("Correlation Data Tail:")
print(corr_data.tail())

# Check if correlation data is empty
if corr_data.empty:
    print("Warning: Correlation data is empty. No plot will be generated.")
else:
    # Plot rolling correlations
    plt.figure(figsize=(12, 6))
    for col in ['US10Y', 'FFR', 'CPI', 'DXY']:
        plt.plot(corr_data['Date'], corr_data[col], label=f'EUR/USD vs {col}')
    plt.title('90-Day Rolling Correlation of EUR/USD Returns with Macro Indicators', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Correlation', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Volatility (30-day rolling standard deviation of EUR/USD returns)
data['EURUSD_Volatility'] = data['EURUSD_Returns'].rolling(window=30).std() * np.sqrt(252)  # Annualized
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['EURUSD_Volatility'], label='EUR/USD Volatility (30-day)', color='red')
plt.title('EUR/USD Annualized Volatility (2006-2025)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Volatility', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Overlay EUR/USD with CPI
fig, ax1 = plt.subplots(figsize=(12, 6))
ax1.plot(data.index, data['EURUSD'], color='blue', label='EUR/USD')
ax1.set_xlabel('Date', fontsize=12)
ax1.set_ylabel('EUR/USD', color='blue', fontsize=12)
ax1.tick_params(axis='y', labelcolor='blue')
ax2 = ax1.twinx()
ax2.plot(data.index, data['CPI'], color='green', label='CPI')
ax2.set_ylabel('CPI', color='green', fontsize=12)
ax2.tick_params(axis='y', labelcolor='green')
plt.title('EUR/USD vs CPI (2006-2025)', fontsize=14)
fig.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=2)
plt.grid(True)
plt.tight_layout()
plt.show()

# 4. Basic Backtest (SMA Crossover Strategy)
# Calculate SMAs
data['SMA_Short'] = data['EURUSD'].rolling(window=20).mean()
data['SMA_Long'] = data['EURUSD'].rolling(window=50).mean()

# Generate signals
data['Signal'] = 0
data['Signal'] = np.where(data['SMA_Short'] > data['SMA_Long'], 1, -1)

# Calculate strategy returns
data['Strategy_Returns'] = data['EURUSD_Returns'] * data['Signal'].shift(1)

# Cumulative returns
data['Cumulative_Strategy'] = (1 + data['Strategy_Returns']).cumprod()
data['Cumulative_BuyHold'] = (1 + data['EURUSD_Returns']).cumprod()

# Plot equity curve
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Cumulative_Strategy'], label='SMA Crossover Strategy', color='purple')
plt.plot(data.index, data['Cumulative_BuyHold'], label='Buy and Hold', color='blue')
plt.title('SMA Crossover Strategy vs Buy and Hold', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Cumulative Returns', fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Risk Metrics
# Annualized Sharpe Ratio
sharpe_ratio = (data['Strategy_Returns'].mean() * 252) / (data['Strategy_Returns'].std() * np.sqrt(252))
print(f"Sharpe Ratio: {sharpe_ratio:.2f}")

# Maximum Drawdown
rolling_max = data['Cumulative_Strategy'].cummax()
drawdown = (data['Cumulative_Strategy'] - rolling_max) / rolling_max
max_drawdown = drawdown.min()
print(f"Maximum Drawdown: {max_drawdown:.2%}")

# 5. Save Data (Optional)
data.to_csv('eurusd_macro_analysis.csv')
print("Data saved to 'eurusd_macro_analysis.csv'")
```

```python
# Install libraries silently
!pip install --quiet numpy==1.26.4 pandas yfinance fredapi statsmodels matplotlib seaborn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from fredapi import Fred
import statsmodels.api as sm

# Set plot style
plt.style.use('seaborn-v0_8')
sns.set_palette("deep")

# Define date range
start_date = '2006-01-01'
end_date = '2024-12-31'

# Initialize FRED API
fred = Fred(api_key='e1020df616b278b14d10a28764a8b2a1')

# Pull EUR/USD from yfinance
try:
    eurusd = yf.download('EURUSD=X', start=start_date, end=end_date, progress=False)
    if eurusd.empty:
        raise ValueError("No EUR/USD data retrieved from yfinance.")
    close_data = eurusd['Close']
    print("Type of eurusd['Close']:", type(close_data))
    if isinstance(close_data, pd.Series):
        eurusd = close_data.to_frame(name='EURUSD')
    elif isinstance(close_data, pd.DataFrame):
        eurusd = close_data.rename(columns={close_data.columns[0]: 'EURUSD'})
    else:
        raise ValueError(f"Unexpected type for eurusd['Close']: {type(close_data)}")
    eurusd = eurusd.loc[:end_date]  # Enforce end date
    print("EUR/USD Data Retrieved:")
    print(eurusd.head())
    print(eurusd.tail())
    print("EUR/USD Columns:", eurusd.columns.tolist())
except Exception as e:
    print(f"Error retrieving EUR/USD data: {e}")
    raise

# Pull macro indicators from FRED
def fetch_fred_series(series_id, name):
    try:
        data = fred.get_series(series_id, start_date=start_date, end_date=end_date)
        print(f"FRED {series_id} Type:", type(data))
        print(f"FRED {series_id} Head:", data.head())
        if isinstance(data, pd.Series):
            data = data.to_frame(name=name)
        elif isinstance(data, pd.DataFrame):
            data = data.rename(columns={data.columns[0]: name})
        else:
            raise ValueError(f"Unexpected data type for {series_id}: {type(data)}")
        data = data.loc[:end_date]  # Enforce end date
        return data
    except Exception as e:
        print(f"Error retrieving FRED series {series_id}: {e}")
        raise

us10y = fetch_fred_series('DGS10', 'US10Y')
ffr = fetch_fred_series('FEDFUNDS', 'FFR')
cpi = fetch_fred_series('CPIAUCSL', 'CPI')
dxy = fetch_fred_series('DTWEXBGS', 'DXY')

# Combine and clean data
data = eurusd.join([us10y, ffr, cpi, dxy], how='outer')
data = data.loc[:end_date].ffill().dropna()  # Enforce end date, use trading days

# Calculate returns
data['EURUSD_Returns'] = data['EURUSD'].pct_change()
data['US10Y_Returns'] = data['US10Y'].pct_change()
data['FFR_Returns'] = data['FFR'].pct_change()
data['CPI_Returns'] = data['CPI'].pct_change()
data['DXY_Returns'] = data['DXY'].pct_change()
data = data.dropna()

# Verify data
print("Data Shape:", data.shape)
print("Data Columns:", data.columns.tolist())
print("Combined Data Tail:")
print(data.tail())

# Rolling correlations (90-day window)
corr_data = pd.DataFrame(index=data.index)
for col in ['US10Y_Returns', 'FFR_Returns', 'CPI_Returns', 'DXY_Returns']:
    corr_data[col] = data['EURUSD_Returns'].rolling(window=90).corr(data[col])
corr_data = corr_data.dropna()
corr_data['Date'] = corr_data.index
print("Correlation Data Shape:", corr_data.shape)
print("Correlation Data Columns:", corr_data.columns.tolist())
print("Correlation Data Head:")
print(corr_data.head())
print("Correlation Data Tail:")
print(corr_data.tail())

# Plot rolling correlations
if not corr_data.empty:
    plt.figure(figsize=(12, 6))
    for col in ['US10Y_Returns', 'FFR_Returns', 'CPI_Returns', 'DXY_Returns']:
        plt.plot(corr_data['Date'], corr_data[col], label=f'EUR/USD vs {col.replace("_Returns", "")}')
    plt.title('90-Day Rolling Correlation of EUR/USD Returns with Macro Factors', fontsize=14)
    plt.xlabel('Date', fontsize=12)
    plt.ylabel('Correlation', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("Warning: No correlation data to plot.")

# Heatmaps of correlation matrices
years = pd.date_range(start='2006-01-01', end='2024-12-31', freq='2YE')
for year in years:
    window_start = year - pd.Timedelta(days=45)
    window_end = year + pd.Timedelta(days=45)
    window_data = data.loc[window_start:window_end, ['EURUSD_Returns', 'US10Y_Returns', 'FFR_Returns', 'CPI_Returns', 'DXY_Returns']]
    if len(window_data) < 10:
        print(f"Skipping heatmap for {year.strftime('%Y-%m-%d')}: insufficient data ({len(window_data)} rows)")
        continue
    corr_matrix = window_data.corr()
    plt.figure(figsize=(8, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0)
    plt.title(f'Correlation Matrix around {year.strftime("%Y-%m-%d")}', fontsize=14)
    plt.tight_layout()
    plt.show()

# OLS Regression
X = data[['US10Y_Returns', 'FFR_Returns', 'CPI_Returns', 'DXY_Returns']]
X = sm.add_constant(X)
y = data['EURUSD_Returns']
model = sm.OLS(y, X, missing='drop').fit()
print("\nOLS Regression Results:")
print(model.summary())

# Save data
data.to_csv('eurusd_macro_correlation_analysis.csv')
print("Data saved to 'eurusd_macro_correlation_analysis.csv'")
```
